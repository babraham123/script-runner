```python
def sql_list(iterable,format="int"):
    sql_list=''
    if format=="string":
        for i in iterable:
            sql_list+="'"+str(i)+"'"+","
        
    elif format=="int":
        for i in iterable:
            sql_list+=str(i)+','
            
    else:
        print 'Incorrect format parameter. Choose string or int.'
        
    sql_list = sql_list[:-1]  #delete last comma
    return sql_list
```

```python
def get_toolbar_imps(object_ids,object_type,days_back=7):
    """
    Function to take list of object ID's to return a Pandas dataframe the following columns:
    
    object_id
    object_name
    toolbar_domain
    toolbar_imps
    
    Parameters
    ==========
    object_ids: this can be a Pandas Series (df['col']) or a string of the comma separated object ID's.
    
    object_type: the type of the object. Can be one of:
    
    member
    publisher
    placement

    days_back: the number of days back from yesterday to run the query for. Default is 7 days.
    """
    
    
    
    #parse input
    if type(object_ids)!= str:
        object_ids=sql_list(object_ids,"str")
    
    #get flagged_tags
    flagged_tags_query="""SELECT 
        b.member_id,
       c.billing_name AS member_name,
       d.id as publisher_id,
       d.name AS publisher_name,
       a.tinytag_id,
       e.name AS tinytag_name,
       a.last_modified
        FROM api.tinytag_inventory_attribute a
        LEFT JOIN bidder.quick_stats b
         ON (b.object_id = a.tinytag_id
        AND b.report_type = 'placement'
        AND b.time_interval = '7day'
        AND a.deleted = 0)
      LEFT JOIN api.tinytag b ON b.id = a.tinytag_id
      LEFT JOIN api.member c ON b.member_id = c.id
      LEFT JOIN api.tinytag e ON e.id = a.tinytag_id
      LEFT JOIN api.publisher d ON e.publisher_id = d.id
    WHERE a.inventory_attribute_id = 17
    AND   a.is_system = TRUE
    AND {} IN ({})
    GROUP BY 1,2,3,4,5,6,7
    HAVING SUM(b.imps) > 0
    """
        
    if object_type=="member":
        flagged_tags=prod.select_dataframe(flagged_tags_query .format("c.id",object_ids))
        
    elif object_type=="publisher":
        flagged_tags=prod.select_dataframe(flagged_tags_query .format("d.id",object_ids))
        
    elif object_type=="placement":
        flagged_tags=prod.select_dataframe(flagged_tags_query .format("e.id",object_ids))
        
    else:
        print "Invalid object_type. Please input member, publisher or placement"
        
        
    #get list of toolbar domains
    toolbar_domains=icq.select_dataframe("""
    SELECT distinct domain from toolbar_domains where deleted=0""")
    
    toolbar_list=sql_list(toolbar_domains.domain,'string')
    flagged_tags_string=sql_list(flagged_tags.tinytag_id,'int')
    
    #get bad domains for domain_visibility metric
    good_domains = prod.select_dataframe("""select audit_status, url from api.inventory_url where audit_status IN ('audited')
    and deleted=0;""")
    
    good_domains_list=sql_list(good_domains['url'],'string')
    toolbar_traffic=vertica.select_dataframe("""
    SELECT a.tag_id,
       CASE
         WHEN a.site_domain IN ({}) THEN a.site_domain
         ELSE 'NA'
       END AS 'toolbar_domain',
       CASE
         WHEN a.site_domain IN ({}) THEN 'Toolbar'
         WHEN a.site_domain IN ({}) THEN 'Audited'
         ELSE 'Masked/Unaudited'
       END AS 'Domain Type',
       SUM(a.imps) imps
        FROM agg_dw_seller_site_domain a
        WHERE a.tag_id IN ({})
        AND   a.ymd BETWEEN CURRENT_DATE- {} AND CURRENT_DATE -1
        AND   a.imp_type != 7
        GROUP BY 1,2,3
    """ .format(toolbar_list,
                toolbar_list,
                good_domains_list,
                flagged_tags_string,
                str(days_back+1)
                ))
    
    df=flagged_tags.merge(toolbar_traffic, left_on='tinytag_id', right_on='tag_id')
    
    return df
```
